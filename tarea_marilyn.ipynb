```python
with open("pubmed-leishmania-set.txt", errors="ignore") as l: 
    texto = l.read()
```

```python
texto = re.sub(r"\n\s{6}", " ", texto)
```

```python
countries_1 = re.findall (r"AD\s{2}-\s[A-Za-z].*,\s([A-Za-z]*)\.\s", texto)
```

```python
countries_2 = re.findall (r"AD\s{2}-\s[A-Za-z].*,\s(\w.*)\.\sElectronic", texto) 
```
